```python
import pandas as pd 
```

```python
# df = pd.read_csv('./Canada-death-causes.csv')
df = pd.read_csv('./Death-Cause-Gender.csv')
```

```python
df['REF_DATE']
```

```python
df.head(3)
df['REF_DATE'].unique()
```

```python
df_clean = df.drop(columns=['UOM_ID','STATUS','DECIMALS','TERMINATED','STATUS','SCALAR_FACTOR','SCALAR_ID','SYMBOL','DGUID','VECTOR','COORDINATE'])
```

```python
df_clean
```

```python
df_clean.rename( columns={'REF_DATE':'Year','GEO':'Province','Sex':'Gender','Leading causes of death (ICD-10)':'Cause of death','UOM':'Metrics','VALUE':'Total'},inplace=True)
```

```python
df_clean
```

```python
# Cleaning up the data
df_clean = df_clean[
    (df_clean['Province'] != 'Canada, place of residence')&
    (
        (df_clean['Metrics'] != 'Percent')&
        (df_clean['Metrics'] != 'Rate per 100,000 population')
    )&
    (df_clean['Gender'] != 'Both sexes')&
    (df_clean['Characteristics'] != 'Rank of leading causes of death')&
    (df_clean['Year'] >= 2018)
    ]
```

```python
df_clean.dropna(subset=['Total'],inplace=True)
```

```python
df_clean['Cause of death'].value_counts()
```

```python
df_clean['Cause of death'] = df_clean['Cause of death'].str.split("[").str[0]
df_clean['Province'] = df_clean['Province'].str.split(",").str[0]
```

```python
df_clean = df_clean[df_clean['Cause of death'] != 'Total, all causes of death ']
```

#### Mapping Province Latitude and longitude coordinations

```python
# Province
Provinces= {
    "Alberta":"53.9333, -116.5765",
    "British Columbia": "53.7267, -127.6476",
    "Manitoba": "49.8951, -97.1384",
    "New Brunswick": "46.5653, -66.4619",
    "Newfoundland and Labrador": "53.1355, -57.6604",
    "Nova Scotia": "44.6820, -63.7443",
    "Ontario": "51.2538, -85.3232",
    "Prince Edward Island": "46.5107, -63.4168",
    "Quebec": "52.9399, -73.5491",
    "Saskatchewan": "52.9399, -106.4509",
    "Northwest Territories": "64.8255, -124.8457",
    "Nunavut": "70.2998, -83.1076",
    "Yukon": "64.2823, -135.0000"
    }
```

```python
Provinces
```

```python
df_clean['coordinate'] = df_clean['Province'].map(Provinces)
# Assign Latitude
df_clean['lat'] = df_clean['coordinate'].str.split(",").str[0]
df_clean['lng'] = df_clean['coordinate'].str.split(",").str[1]
df_clean['Country'] = 'Canada'
```
